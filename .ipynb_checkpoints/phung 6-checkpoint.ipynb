```python
from fastai.vision.all import *
import torch
```

```python
# Download and prepare the CIFAR10 dataset
path = untar_data(URLs.CIFAR)

path
```

```python

dls = ImageDataLoaders.from_folder(
    path, 
    seed=42, 
    valid_pct=0.2, 
    bs=16, 
    device=torch.device('cuda'), 
    item_tfms=Resize(224), 
    batch_tfms=aug_transforms(size=224),
)
```

```python
dls.show_batch()
# Get the number of batches in the training and validation 
num_train_batches = len(dls.train) 
num_valid_batches = len(dls.valid) 
print(f"Number of training batches: {num_train_batches}")
print(f"Number of validation batches: {num_valid_batches}") 
# Get the shape of a single batch (images and labels) from the training 
train_batch = dls.train.one_batch() 
train_images, train_labels = train_batch 
print(f"Shape of training images batch: {train_images.shape}")
print(f"Shape of training labels batch: {train_labels.shape}")
# Get the shape of a single batch (images and labels) from the validation 
valid_batch = dls.valid.one_batch() 
valid_images, valid_labels = valid_batch 
print(f"Shape of validation images batch: {valid_images.shape}")
print(f"Shape of validation labels batch: {valid_labels.shape}") 
# Print dataset size 
num_train_samples = len(dls.train_ds) 
num_valid_samples = len(dls.valid_ds) 
print(f"Number of training samples: {num_train_samples}")
print(f"Number of validation samples: {num_valid_samples}")
```

```python
learn = vision_learner(
    dls, 
    resnet50, 
    metrics=[error_rate, accuracy],
    # loss_func=torch.nn.BCELoss(),
    wd=0.02,
    lr=5e-3,
    cbs=[
        ReduceLROnPlateau(min_delta=0.01, patience=2),
        EarlyStoppingCallback(min_delta=0.01, patience=3),
        SaveModelCallback(fname='cifar10_resnet50', at_end=True)
    ]
)
```

```python
learn.lr_find()
```

```python
learn.fine_tune(15, 1e-3)
```

```python
# Save the trained model
learn.export('cifar10_resnet50_202405302045.pkl')
```

```python
interpretation = Interpretation.from_learner(learn)
```

```python
interpretation.plot_top_losses(10)
```
