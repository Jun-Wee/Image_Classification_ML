```python
from fastai.vision.all import *
import torch
```

```python
# Download and prepare the CIFAR10 dataset
path = untar_data(URLs.CIFAR)

path.ls()

print(len((path/'train').ls()), len((path/'test').ls()))
for folder in (path/'train').ls():
    print(folder, len(folder.ls()))
print('\n')
for folder in (path/'test').ls():
    print(folder, len(folder.ls()))
```

```python

dls = ImageDataLoaders.from_folder(
    (path/'train'), 
    seed=42, 
    valid_pct=0.3, 
    bs=32, 
    device=torch.device('cuda'), 
    item_tfms=Resize(224), 
    batch_tfms=aug_transforms(size=224),
)

dls.valid_ds
```

```python
print('Number of training samples:', len(dls.train_ds), '\nNumber of validation samples:', len(dls.valid_ds))
```

```python
dls.show_batch()
```

```python
learn = vision_learner(
    dls, 
    resnet50, 
    metrics=[error_rate, accuracy],
    path="./models",
    wd=0.02,
    lr=5e-3,
    cbs=[
        ReduceLROnPlateau(min_delta=0.01, patience=2),
        EarlyStoppingCallback(min_delta=0.001, patience=3),
        SaveModelCallback(fname='checkpoint_cifar10_resnet50', with_opt=True),
        CSVLogger(fname='history.csv', append=False),
        ShowGraphCallback(),
    ]
)
```

```python
learn.lr_find()
```

```python
learn.fine_tune(15, 1e-3)
```

```python
import dill
MODEL_VERSION = '0.2.1'
MODEL_PATH = 'cifar10_resnet50_v_11_2'
learn.remove_cb(CSVLogger)
learn.export(MODEL_PATH)
```

```python
path = untar_data(URLs.CIFAR)
test_files = get_image_files(path/'test')
# resize to 224
test_dl = learn.dls.test_dl(
    test_files,
    with_labels=True,
    device=torch.device('cuda'),
    item_tfms=Resize(224),
    batch_tfms=aug_transforms(size=224)
)

test_dl.show_batch()
# show resolution
print(test_dl.one_batch()[0].shape)
```

```python
classification_interpretation = ClassificationInterpretation.from_learner(learn, dl=test_dl)
```

```python
classification_interpretation.plot_confusion_matrix(title='Confusion matrix')
classification_interpretation.plot_confusion_matrix(title='Confusion with accuracy matrix', normalize=True)
```

```python
classification_interpretation.plot_top_losses(50)
```

```python
@patch
@delegates(subplots)
def plot_metrics(self: Recorder, nrows=None, ncols=None, figsize=None, **kwargs):
    metrics = np.stack(self.values)
    names = self.metric_names[1:-1]
    n = len(names) - 1
    if nrows is None and ncols is None:
        nrows = int(math.sqrt(n))
        ncols = int(np.ceil(n / nrows))
    elif nrows is None: nrows = int(np.ceil(n / ncols))
    elif ncols is None: ncols = int(np.ceil(n / nrows))
    figsize = figsize or (ncols * 6, nrows * 4)
    fig, axs = subplots(nrows, ncols, figsize=figsize, **kwargs)
    axs = [ax if i < n else ax.set_axis_off() for i, ax in enumerate(axs.flatten())][:n]
    for i, (name, ax) in enumerate(zip(names, [axs[0]] + axs)):
        ax.plot(metrics[:, i], color='#1f77b4' if i == 0 else '#ff7f0e', label='valid' if i > 0 else 'train')
        ax.set_title(name if i > 1 else 'losses')
        ax.legend(loc='best')
    plt.show()
```

```python
learn.recorder.plot_metrics()
```
