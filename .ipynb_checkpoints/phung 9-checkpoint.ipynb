```python
from fastai.vision.all import *
import torch
```

```python
# Download and prepare the CIFAR10 dataset
path = untar_data(URLs.CIFAR)

path.ls()

print(len((path/'train').ls()), len((path/'test').ls()))
for folder in (path/'train').ls():
    print(folder, len(folder.ls()))
print('\n')
for folder in (path/'test').ls():
    print(folder, len(folder.ls()))
```

```python

dls = ImageDataLoaders.from_folder(
    path, 
    seed=42, 
    valid_pct=0.2, 
    bs=16, 
    device=torch.device('cuda'), 
    item_tfms=Resize(224), 
    batch_tfms=aug_transforms(size=224),
)
```

```python
print('Number of training samples:', len(dls.train_ds), '\nNumber of validation samples:', len(dls.valid_ds))
```

```python
dls.show_batch()
```

```python
learn = vision_learner(
    dls, 
    resnet50, 
    metrics=[error_rate, accuracy],
    # loss_func=torch.nn.BCELoss(),
    wd=0.02,
    lr=5e-3,
    cbs=[
        ReduceLROnPlateau(min_delta=0.01, patience=2),
        EarlyStoppingCallback(min_delta=0.001, patience=3),
        SaveModelCallback(fname='cifar10_resnet50', at_end=True)
    ]
)
```

```python
learn.lr_find()
```

```python
learn.fine_tune(15, 1e-3)
```

```python
# Save the trained model
learn.export('cifar10_resnet50_202406012038.pkl')
```

```python
interpretation = Interpretation.from_learner(learn)
```

```python
interpretation.plot_top_losses(32)
```
